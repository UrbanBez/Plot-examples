```python
# Import packages

%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import ticker
from matplotlib import rc
import seaborn as sns

# Set the global font to be Arial, size 10
rc('font',**{'family':'sans-serif','sans-serif':['Arial'],'size':10})
```

```python
# Import data

data_fig1b = pd.read_csv("datasets/data_fig1b.csv")
```

```python
#### PLOT ####

#################################

# Define plot style. I use seaborn's ticks style and colorblind colormap. 
# See http://seaborn.pydata.org/tutorial/aesthetics.html#seaborn-figure-styles 
# and http://seaborn.pydata.org/tutorial/color_palettes.html#qualitative-color-palettes

sns.set(style="ticks", palette="colorblind")

# Plot the colormap (uncomment to view).

#sns.palplot(sns.color_palette()) #uncomment

# Set figure properties.

fig = plt.figure(figsize=(3*1.618,3), dpi = 300) # golden ratio
ax1 = fig.add_subplot(111)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
ax1.tick_params('both', length=6, width=1.2, which='major')
ax1.tick_params('both', length=4, width=1.2, which='minor')

for axis in ['top','bottom','left','right']:
    ax1.spines[axis].set_linewidth(1.2)

# Define the color palette. This creates 10 distinct colors from the "colorblind" colormap (index 0-9).    

palette = sns.color_palette(n_colors=10)

# Plot the data.

plt.fill_between(data_fig1b['time'], data_fig1b['mean'] - data_fig1b['std'], data_fig1b['mean'] + data_fig1b['std'], color=palette[0], alpha=0.5) # +/- SD as shaded area
ax1.plot(data_fig1b['time'], data_fig1b['mean'], linewidth=1.6, alpha=1)

# Define y range.

ax1.set_ylim([0, 1.05])

# Set the axis and tick properties

ax1.autoscale(enable=True, axis='x', tight=True)
ax1.xaxis.set_ticks_position('bottom')
ax1.yaxis.set_ticks_position('left')
locx = ticker.MultipleLocator(base=15) # this locator puts ticks at regular intervals
ax1.xaxis.set_major_locator(locx)
locy = ticker.MultipleLocator(base=0.25) # this locator puts ticks at regular intervals
ax1.yaxis.set_major_locator(locy)

# Label the axis.

ax1.set_xlabel("time (min)", fontsize=15)
ax1.set_ylabel("normalized intensity", fontsize=15)

# Set the font for .svg export

plt.rcParams.update({'mathtext.default':  'regular', "svg.fonttype": 'none' })

sns.despine()

# Save the figure

# plt.savefig('fig1B_example.png', format='png', dpi = 150, bbox_inches='tight',edgecolor='none')
# plt.savefig('fig1B_example.svg', format='svg', bbox_inches='tight',edgecolor='none')
```
