```python
# Import packages

%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import ticker
from matplotlib import rc
import seaborn as sns
from scipy.optimize import curve_fit
import scipy.stats

# Set the global font to be Arial, size 10
rc('font',**{'family':'sans-serif','sans-serif':['Arial'],'size':10})
```

```python
 # Import data

import tkinter as tk
from tkinter import filedialog

root= tk.Tk()

canvas1 = tk.Canvas(root, width = 320, height = 200, bg = 'lightsteelblue2', relief = 'raised')
canvas1.pack()

def getCSV_data ():
    global df
    
    import_file_path = filedialog.askopenfilename()
    df = pd.read_csv (import_file_path)
    df = df.drop(df.columns[0], axis=1)
    display(df)
    
browseButton_CSV = tk.Button(text="      Import grid data CSV File     ", command=getCSV_data, bg='green', fg='white', font=('arial', 12, 'bold'))
canvas1.create_window(160, 100, window=browseButton_CSV)

root.mainloop()


data_grids = df
```

```python
 # Import data

root= tk.Tk()

canvas1 = tk.Canvas(root, width = 320, height = 200, bg = 'oldlace', relief = 'raised')
canvas1.pack()

def getCSV_noise ():
    global df
    
    import_file_path = filedialog.askopenfilename()
    df = pd.read_csv (import_file_path)
    df = df.drop(df.columns[0], axis=1)
    display(df)
        
    
browseButton_CSV = tk.Button(text="      Import noise data CSV File     ", command=getCSV_noise, bg='red', fg='white', font=('arial', 12, 'bold'))
canvas1.create_window(160, 100, window=browseButton_CSV)

root.mainloop()


data_noise = df
```

```python
# Subtract the noise

data_grids_sub = data_grids-data_noise.iloc[0,0]

# Time column

time = [t*0.5 for t in range(0,data_grids_sub.shape[0])]
data_grids_sub['Time'] = time

display(data_grids_sub)
```

```python
#### PLOT ####

from pandas import *

data_grids_sub_means = data_grids_sub.filter(regex='Mean')

#################################

sns.set(style="ticks", palette="colorblind")

fig = plt.figure(figsize=(3*1.618,3), dpi = 300) # golden ratio
ax1 = fig.add_subplot(111)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
ax1.tick_params('both', length=6, width=1.2, which='major')
ax1.tick_params('both', length=4, width=1.2, which='minor')

for axis in ['top','bottom','left','right']:
    ax1.spines[axis].set_linewidth(1.2)    

cm = plt.get_cmap('plasma')
ax1.set_prop_cycle('color',[cm(1.*i/data_grids_sub_means.shape[1]) for i in range(data_grids_sub_means.shape[1])])

# Plot the data.

for i in range(0,data_grids_sub_means.shape[1]):
    ax1.plot(data_grids_sub['Time'], data_grids_sub_means.iloc[:,i], linewidth=1.6, alpha=0.7)

# Define y range.

ax1.set_ylim([100, 8e3])

# Set the axis and tick properties

ax1.autoscale(enable=True, axis='x', tight=True)
ax1.xaxis.set_ticks_position('bottom')
ax1.yaxis.set_ticks_position('left')
locx = ticker.MultipleLocator(base=30) # this locator puts ticks at regular intervals
ax1.xaxis.set_major_locator(locx)
locy = ticker.MultipleLocator(base=2e3) # this locator puts ticks at regular intervals
ax1.yaxis.set_major_locator(locy)

# Label the axis.

ax1.set_xlabel("time (min)", fontsize=15)
ax1.set_ylabel("fluorescence intensity", fontsize=15)

# Set the font for .svg export

plt.rcParams.update({'mathtext.default':  'regular', "svg.fonttype": 'none' })

sns.despine()

# Save the figure

# plt.savefig('200615_100nM_timecourses.png', format='png', dpi = 300, bbox_inches='tight',edgecolor='none')
# plt.savefig('fig1B_example.svg', format='svg', bbox_inches='tight',edgecolor='none')
```

```python
# Filter activated compartments with intensity cutoff

k = 1000

data_grids_sub_means_active = pd.DataFrame()

for i in range(data_grids_sub_means.shape[1]):
    if data_grids_sub_means.iloc[-1,i]>k:
        data_grids_sub_means_active=data_grids_sub_means_active.append(data_grids_sub_means.iloc[:,i])
data_grids_sub_means_active=data_grids_sub_means_active.transpose()       
display(data_grids_sub_means_active)
```

```python
# Do a Gompertz fit on active compartments

grid_Gomp_params = pd.DataFrame(columns=data_grids_sub_means_active.columns)
grid_Gomp_errs = pd.DataFrame(columns=data_grids_sub_means_active.columns)

def gomp(x, A, B, Ti, ku):
    return B + (A - B) * np.exp(-np.exp(- np.e*ku * (x - Ti)))

B = 0
A = 6000
Ti = 15
ku = 0.01
guess = np.array([A, B, Ti, ku])

for i in range(data_grids_sub_means_active.shape[1]):
    try:
        popt, pcov = curve_fit(gomp, data_grids_sub['Time'], data_grids_sub_means_active.iloc[:,i], guess, bounds = ((0, -np.inf, 0, 0), (10e4, np.inf, 128.5, 0.2)))
        perr = np.sqrt(np.diag(pcov))
        grid_Gomp_params.iloc[:,i] = popt
        grid_Gomp_errs.iloc[:,i] = perr
    except RuntimeError:
        print("Error - curve_fit failed",data_grids_sub_means_active.columns[i])
        grid_Gomp_params.iloc[:,i] = np.nan
        grid_Gomp_errs.iloc[:,i] = np.nan
    
display(grid_Gomp_params) 
display(grid_Gomp_errs) 
```

```python
# Display parameter histograms

#### PLOT ####

sns.set(style="ticks", palette="colorblind")

fig = plt.figure(figsize=(3*1.618,3), dpi = 300) # golden ratio
ax1 = fig.add_subplot(111)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
ax1.tick_params('both', length=6, width=1.2, which='major')
ax1.tick_params('both', length=4, width=1.2, which='minor')

for axis in ['top','bottom','left','right']:
    ax1.spines[axis].set_linewidth(1.2)    

palette = sns.color_palette(n_colors=10)

# Plot the data.

sns.distplot(grid_Gomp_params.iloc[3],fit=scipy.stats.lognorm, kde=False,norm_hist=True)
norm_params = scipy.stats.lognorm.fit(grid_Gomp_params.iloc[3].dropna())
print('Gaussian parameters', norm_params)

# Define y range.

# ax1.set_ylim([200, 8e3])

# Set the axis and tick properties

ax1.autoscale(enable=True, axis='x', tight=True)
ax1.xaxis.set_ticks_position('bottom')
ax1.yaxis.set_ticks_position('left')

# Label the axis.

ax1.set_xlabel("$\it{k_{max}}$ $(min^{-1})$", fontsize=15)
ax1.set_ylabel("density", fontsize=15)

# Set the font for .svg export

plt.rcParams.update({'mathtext.default':  'regular', "svg.fonttype": 'none' })

sns.despine()

# Save the figure

# plt.savefig('200615_100nM_kmax_hist.png', format='png', dpi = 150, bbox_inches='tight',edgecolor='none')
# plt.savefig('fig1B_example.svg', format='svg', bbox_inches='tight',edgecolor='none')
```
