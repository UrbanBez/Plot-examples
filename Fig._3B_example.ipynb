```python
# Import packages

%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import ticker
from matplotlib import rc
import seaborn as sns

# Set the global font to be Arial, size 10
rc('font',**{'family':'sans-serif','sans-serif':['Arial'],'size':10})
```

```python
# Import data

data_GTP = pd.read_csv("Rab5GTP_data.csv")
data_GDP = pd.read_csv("Rab5GDP_data.csv")
```

```python
#### PLOT ####

#################################

# Define plot style. I use seaborn's ticks style and colorblind colormap. 
# See http://seaborn.pydata.org/tutorial/aesthetics.html#seaborn-figure-styles 
# and http://seaborn.pydata.org/tutorial/color_palettes.html#qualitative-color-palettes

sns.set(style="ticks", palette="colorblind")

# Set figure properties.

fig = plt.figure(figsize=(2*1.618,2), dpi = 300) # golden ratio
ax1 = fig.add_subplot(111)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
ax1.tick_params('both', length=6, width=1.2, which='major')
ax1.tick_params('both', length=4, width=0, which='minor')

for axis in ['top','bottom','left','right']:
      ax1.spines[axis].set_linewidth(1.2)
        
# For this plot I use custom colors to match particle trajectories in Fig. 3B.

colorGTP = plt.cm.viridis(0.63)
colorGDP = plt.cm.viridis(0.2)

# Plot the data.

ax1.scatter(data_GDP['bins'][data_GDP['bins']<3],data_GDP['freq'][data_GDP['bins']<3], color = colorGDP, s=25, alpha=0.6)
ax1.scatter(data_GTP['bins'][data_GTP['bins']<22],data_GTP['freq'][data_GTP['bins']<22], color = colorGTP, s=25, alpha=0.6)
ax1.legend(bbox_to_anchor=(1.0, 1.3),loc = 1, fontsize=15, frameon=False, labels=['Rab5:GDP', 'Rab5:GTP'])

# Format the ticks and axis

ax1.xaxis.set_ticks_position('bottom')
ax1.yaxis.set_ticks_position('left')
locx = ticker.MultipleLocator(base=10) # this locator puts ticks at regular intervals
ax1.xaxis.set_major_locator(locx)

ax1.set_yscale('log')
plt.yticks([1e0,1e-2,1e-4])

# Label the axis

ax1.set_xlabel("track duration (s)", fontsize=15)
ax1.set_ylabel("frequency", fontsize=15)

plt.rcParams.update({'mathtext.default':  'regular', "svg.fonttype": 'none' }) # For .svg font editing

sns.despine()
```
